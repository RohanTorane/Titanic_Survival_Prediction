```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv('../../Datasets/Titanic/train.csv')
df_train.head()
```

```python
df_train.shape
```

```python
df_train.columns
```

```python
selected_columns = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
```

## Handling of missing values

```python
#number of missing values
df_train.isna().sum()
```

```python
df_train.info()
```

```python
df_train['Age'].fillna(df_train['Age'].mean(),inplace=True)
df_train['Embarked'].fillna(df_train['Embarked'].mode().iloc[0],inplace=True)
```

```python
df_train.isna().sum()
```

```python
df_train.info()
```

```python
df_train.head()
```

```python
df_train['Sex'] = df_train['Sex'].map({'male':0,'female':1})
df_train['Embarked'] = df_train['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
df_train
```

```python
X = df_train[selected_columns]
y = df_train['Survived']
```

```python
X.isna().sum()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
```

# GridSearchCV_RF

```python
parametrs = {'bootstrap':[True,False],'max_features':['auto','sqrt'],'min_samples_split':[2,5],'max_depth':[2,3,4,5], 'min_samples_leaf':[1,2,3,4,5,6],'n_jobs':[-1]}
grid_cv_rf = GridSearchCV(RandomForestClassifier(random_state=100), param_grid=parametrs)
grid_cv_rf.fit(X,y)
```

```python
grid_cv_rf.best_score_
```

```python
parametrs = {'bootstrap':[True,False],'class_weight':[None], 'criterion':['entropy'],'max_depth':[2,3,4,5], 'max_features':['auto','sqrt'], 
             'max_leaf_nodes':[None],'min_impurity_decrease':[0.0],
             'min_samples_leaf':[1,2,3,4,5,6], 'min_samples_split':[2,5],'min_weight_fraction_leaf':[0.0], 'n_estimators':[50],
            'n_jobs': [-1], 'oob_score':[False], 'random_state' : [100] ,'warm_start':[False]}
grid_cv_rf = GridSearchCV(RandomForestClassifier(random_state=100), param_grid=parametrs)
grid_cv_rf.fit(X,y)
```

```python
df_test = pd.read_csv('../../Datasets/Titanic/test.csv')
df_test.head()
```

```python
df_test.isna().sum()
```

```python
df_test['Age'].fillna(df_train['Age'].mean(),inplace=True)
df_test.head()
```

```python
df_test['Fare'].fillna(df_train['Fare'].median(),inplace=True)
df_test.isna().sum()
```

```python
selected_columns
```

```python
df_test = df_test[selected_columns]
df_test['Sex'] = df_test['Sex'].map({'male':0,'female':1})
df_test['Embarked'] = df_test['Embarked'].map({'S':0,'C':1,'Q':2})
df_test.head()
```

```python
#Gridsearch CV
yp = grid_cv_rf.predict(df_test)
df_test = pd.read_csv('../../Datasets/Titanic/test.csv')
df_test['Survived'] = yp
df_test[['PassengerId','Survived']].to_csv('submission_survived.csv',index=False)
```
